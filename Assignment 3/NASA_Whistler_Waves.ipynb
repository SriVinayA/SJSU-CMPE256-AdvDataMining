```python
import librosa
import librosa.display
import matplotlib.pyplot as plt
```

```python
audio_path = '/content/drive/MyDrive/SJSU/Sem1/CMPE256 - Advanced Data Mining/Assign3/whistler.mp3'  # Replace with your audio file path
y, sr = librosa.load(audio_path)
```

```python
plt.figure(figsize=(10, 4))
D = librosa.amplitude_to_db(np.abs(librosa.stft(y)), ref=np.max)
librosa.display.specshow(D, sr=sr, x_axis='time', y_axis='log')
plt.colorbar(format='%+2.0f dB')
plt.title('Spectrogram')
plt.tight_layout()
plt.show()
```

Displayed a spectrogram showing the distribution of frequencies over time, with color intensity indicating amplitude.

```python
# Compute a mel-scaled spectrogram
S = librosa.feature.melspectrogram(y=y, sr=sr)
plt.figure(figsize=(10, 4))
S_DB = librosa.amplitude_to_db(S, ref=np.max)
librosa.display.specshow(S_DB, sr=sr, x_axis='time', y_axis='mel')
plt.colorbar(format='%+2.0f dB')
plt.title('Mel-frequency spectrogram')
plt.tight_layout()
plt.show()
```

Visualized a mel-frequency spectrogram, emphasizing frequencies in a way that mirrors human auditory perception.

```python
D = librosa.stft(y)
plt.figure(figsize=(10, 4))
librosa.display.specshow(librosa.amplitude_to_db(D, ref=np.max), sr=sr, x_axis='time', y_axis='log')
plt.colorbar(format='%+2.0f dB')
plt.title('STFT Magnitude')
plt.tight_layout()
plt.show()
```

Computed the STFT magnitude and visualized it, highlighting the spectral content over time without phase information.

```python
# Beat tracker
tempo, beats = librosa.beat.beat_track(y=y, sr=sr)
print('Estimated tempo: {:.2f} beats per minute'.format(tempo))

# Plot the beat locations over the waveform
plt.figure(figsize=(14, 5))
librosa.display.waveshow(y, sr=sr, alpha=0.5)
plt.vlines(librosa.frames_to_time(beats, sr=sr), -1, 1, color='r', alpha=0.5, linestyle='--', label='Beats')
plt.legend()
plt.title('Beat tracking')
plt.tight_layout()
plt.show()
```

Estimated the tempo at 143.55 BPM and visualized beat locations superimposed on the audio waveform.

```python
C = librosa.cqt(y, sr=sr)
plt.figure(figsize=(10, 4))
librosa.display.specshow(librosa.amplitude_to_db(C, ref=np.max), sr=sr, x_axis='time', y_axis='cqt_note')
plt.colorbar(format='%+2.0f dB')
plt.title('Constant-Q power spectrum')
plt.tight_layout()
plt.show()
```

Displayed the Constant-Q power spectrum, offering a logarithmic frequency perspective suitable for musical analysis.
