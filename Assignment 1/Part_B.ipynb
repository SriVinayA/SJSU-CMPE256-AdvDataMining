```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage

# Define the data points
data_points = np.array([
    [2, 2], [2, 6], [3, 7], [5, 2], [5, 5],
    [5, 8], [6, 6], [7, 3], [8, 4], [10, 6], [12, 8]
])
```

```python
# Perform hierarchical clustering using different linkage methods
linkage_methods = ['single', 'complete', 'average']
linkage_matrices = {method: linkage(data_points, method=method) for method in linkage_methods}
```

```python
# Plot the dendrograms for each linkage method
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

for ax, (method, matrix) in zip(axes, linkage_matrices.items()):
    dendrogram(matrix, ax=ax)
    ax.set_title(f'{method.capitalize()}-link Clustering')
    ax.set_xlabel('Data points')
    ax.set_ylabel('Distance')

plt.tight_layout()
plt.show()
```
